```python
import turicreate as tc
```

```python
style = tc.load_images('./style')
content = tc.load_images('./content')
```

```python
tc.config.set_num_gpus(-1)
```

```python
model = tc.style_transfer.create(style, content, max_iterations=1000)
```

```python
model.save('KlimtAdeleBlochBauer.model')
```

```python
print(model.summary)
```

```python
model.export_coreml('KlimtAdeleBlochBauer.mlmodel')
```
