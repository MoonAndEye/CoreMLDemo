```python
import turicreate as tc
```

```python
style = tc.load_images('./style')
content = tc.load_images('./content')
```

```python
model = tc.style_transfer.create(style, content, max_iterations=2000)
```

```python
model.save('WWDCDemo.model')
```

```python
model.export_coreml('WWDCDemo.mlmodel')
```

```python
test = content
stylized_image = model.stylize(test, style=1)
```

```python
stylized_image.explore()
```
