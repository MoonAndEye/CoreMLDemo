```python
import turicreate as tc
```

```python
style = tc.load_images('./style')
content = tc.load_images('./content')
```

```python
model = tc.style_transfer.create(style, content, max_iterations=1000)
```

```python
model.save('Munch.model')
```

```python
model.export_coreml('Munch.mlmodel')
```

```python
test = content
stylized_image = model.stylize(test, style=0)
stylized_image.explore()
```
