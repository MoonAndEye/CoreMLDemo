```python
import turicreate as tc
```

```python
style = tc.load_images('./style')
content = tc.load_images('./content')
```

```python
model = tc.style_transfer.create(style, content, max_iterations=1000)
```

```python
model.save('Monet.model')
```

```python
model.export_coreml('Monet.mlmodel')
```
